```python
from tensorflow.keras.utils import load_img, img_to_array,array_to_img
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = load_img('C:\\Users\\user.LAPTOP\\Desktop\\Deep Learning\\CNN\\Data Augmentation\\Elephant.jpg',target_size=(200,200))
```

```python
img
```

```python
data_gen = ImageDataGenerator(rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.2,zoom_range=0.2,horizontal_flip=True,fill_mode='reflect')
```

```python
x = img_to_array(img)
```

```python
type(x)
```

```python
x.shape
```

```python
input_df = x.reshape(1,200,200,3)
```

```python
i = 1

for output in data_gen.flow(input_df,batch_size=1,save_to_dir='C:\\Users\\user.LAPTOP\\Desktop\\Deep Learning\\CNN\\dic',save_prefix='Aug_Elephant'):
    
    i += 1
    
    if i == 21:
        break
```
